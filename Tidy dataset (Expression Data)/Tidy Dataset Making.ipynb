```python
# Requirements 
import numpy as np
import pandas as pd
import os
import re
import tqdm
```

```python
paths = []

for path, root, filename in os.walk('./TCGA-LUAD/Data/gdc_download_20220610_214806.951422/'):
    for file in filename:
        #print (path)
        #print (filename) #Name of each files in TCGA project
        #print (root)
        if re.findall(r'counts', file):  # Just for trimming the annotation files mixed
            paths.append(path + '/' + file)
            #print (path + file)
   
```

```python
len(paths) #Size Of Samples
```

```python
for index, path in tqdm.tqdm (enumerate(paths)):
    if index == 0 :
        df = pd.read_csv(path, sep = '\t', header = 1)
        df = df.drop(range(4))
        df = df[['gene_id', 'gene_name']].set_index('gene_id')
    ex = pd.read_csv(path, sep = '\t', header = 1) 
    ex = ex.drop(range(4))
    ex = pd.DataFrame(ex['unstranded']).values
    colname = re.findall(r'\/.*\/(.*)',path)[0]
    df[colname] = ex
     
```
